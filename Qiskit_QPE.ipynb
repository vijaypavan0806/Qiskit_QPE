```
# Quantum Phase Estimation (QPE) – Complete Version
# Compatible with Qiskit 1.x / Colab / Local Python
# Author: Vijay Pavan Adimulam

# -------------------------------------------------------------------------
# Install any missing dependencies (runs safely even if already installed)
# -------------------------------------------------------------------------
try:
    from qiskit import QuantumCircuit, transpile
    from qiskit.visualization import plot_histogram
    from qiskit_aer import AerSimulator
    from qiskit_aer.noise import NoiseModel, depolarizing_error
    import matplotlib.pyplot as plt
    import numpy as np
except ImportError:
    import sys
    !{sys.executable} -m pip install qiskit qiskit-aer matplotlib pylatexenc --quiet
    from qiskit import QuantumCircuit, transpile
    from qiskit.visualization import plot_histogram
    from qiskit_aer import AerSimulator
    from qiskit_aer.noise import NoiseModel, depolarizing_error
    import matplotlib.pyplot as plt
    import numpy as np


# -------------------------------------------------------------------------
# Function: Inverse Quantum Fourier Transform (Inverse QFT)
# -------------------------------------------------------------------------
def inverse_qft(n):
    """Constructs the inverse QFT on n qubits."""
    qc = QuantumCircuit(n)
    for j in range(n // 2):
        qc.swap(j, n - j - 1)
    for j in range(n):
        for m in range(j):
            qc.cp(-np.pi / float(2 ** (j - m)), m, j)
        qc.h(j)
    qc.name = "Inverse QFT"
    return qc


# -------------------------------------------------------------------------
# Function: QPE Experiment
# -------------------------------------------------------------------------
def qpe_experiment(theta, n_count, noisy=False):
    """
    Perform Quantum Phase Estimation for a given phase theta.
    :param theta: Phase value (0 ≤ θ < 1)
    :param n_count: Number of counting qubits
    :param noisy: Adds noise if True
    """

    # Quantum circuit: n_count counting qubits + 1 target
    qc = QuantumCircuit(n_count + 1, n_count)

    # Step 1: Apply Hadamards to counting qubits
    for q in range(n_count):
        qc.h(q)

    # Step 2: Prepare eigenstate |1⟩
    qc.x(n_count)

    # Step 3: Apply controlled-U(phase) operations
    for q in range(n_count):
        qc.cp(2 * np.pi * theta * (2 ** q), q, n_count)

    # Step 4: Apply inverse QFT
    qc.append(inverse_qft(n_count), range(n_count))

    # ---------------------------------------------------------------------
    # Circuit visualization
    # ---------------------------------------------------------------------
    print(f"\n--- Quantum Circuit for θ={theta}, Counting Qubits={n_count}, Noisy={noisy} ---")
    try:
        qc.draw("mpl")
        plt.show()
    except Exception as e:
        print("(Fallback to text view due to missing 'pylatexenc' or GUI backend)")
        print(qc.draw("text"))

    # Step 5: Measurement
    qc.measure(range(n_count), range(n_count))

    # Step 6: Setup Aer simulator
    simulator = AerSimulator()

    if noisy:
        # Add simple depolarizing noise
        noise_model = NoiseModel()
        noise_model.add_all_qubit_quantum_error(depolarizing_error(0.01, 1), ['u3', 'u2'])
        noise_model.add_all_qubit_quantum_error(depolarizing_error(0.02, 2), ['cx'])
        compiled = transpile(qc, simulator)
        result = simulator.run(compiled, shots=2048, noise_model=noise_model).result()
    else:
        compiled = transpile(qc, simulator)
        result = simulator.run(compiled, shots=2048).result()

    counts = result.get_counts()

    # Step 7: Plot histogram
    plot_histogram(counts)
    plt.title(f"QPE Output Histogram (θ={theta}, {n_count} qubits, noisy={noisy})")
    plt.show()

    # Step 8: Extract measured phase
    measured_bin = max(counts, key=counts.get)
    measured_dec = int(measured_bin, 2) / (2 ** n_count)

    # Step 9: Print comparison
    print(f"Theoretical Phase = {theta:.4f}")
    print(f"Measured Binary   = {measured_bin}")
    print(f"Measured Decimal  = {measured_dec:.4f}")
    print(f"Error             = {abs(theta - measured_dec):.6f}\n")

    return counts, measured_dec


# -------------------------------------------------------------------------
# 🧪 Main Experiment Runner
# -------------------------------------------------------------------------
if __name__ == "__main__":
    phase_values = [0.25, 0.375, 0.5]
    qubit_counts = [3, 4, 5]

    for theta in phase_values:
        for n in qubit_counts:
            qpe_experiment(theta, n, noisy=False)
            qpe_experiment(theta, n, noisy=True)
```
